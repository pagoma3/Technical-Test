```python
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.select import Select

import numpy as np

options = Options()
options.headless = False
options.page_load_strategy = 'normal'

DRIVER_PATH= r'/Users/pau/Downloads/chromedriver'
```

```python
# Open Chrome session and ensure it closes by using "with":
with Chrome(options=options, executable_path=DRIVER_PATH) as driver:
    
    driver.implicitly_wait(15)
    driver.get("https://datosclima.es/Aemethistorico/Lluviasol.php")
    
    # First, we select our province:
    province_sel = driver.find_element_by_name("Provincia")
    Select(province_sel).select_by_visible_text('BARCELONA')
    
    # Then, the station from which measurments have been taken. Specifically, that which has measured since 1920:
    station_sel = driver.find_element_by_name("id_hija")
    Select(station_sel).select_by_visible_text('BARCELONA, FABRA')
    
    # Fill in date forms to achieve the maximum range of data:
    start_day = driver.find_element_by_name("Iday")
    start_day.send_keys("01")
    start_month = driver.find_element_by_name("Imonth")
    start_month.send_keys("01")
    start_year = driver.find_element_by_name("Iyear")
    start_year.send_keys("1970")
    
    end_day = driver.find_element_by_name("Fday")
    end_day.send_keys("31")
    end_month = driver.find_element_by_name("Fmonth")
    end_month.send_keys("07")
    end_year = driver.find_element_by_name("Fyear")
    end_year.send_keys("2021")
    
    # Press button:
    search_button = driver.find_element_by_xpath("//*[@id=\"col2\"]/div[1]/div/form/input[5]")
    search_button.click()

    # Get elements:
    elements = []
    first_table = driver.find_element_by_xpath("//*[@id=\"col2\"]/div/div/table[2]/tbody/tr")
    for foo in first_table.find_elements_by_xpath("./child::*"):
        elements.append(foo.text)
        
    rest_tables = driver.find_element_by_xpath("//*[@id=\"col2\"]/div/div/div[2]")
    for foo in rest_tables.find_elements_by_xpath("./child::*"):
        elements.append(foo.text)
    
```
